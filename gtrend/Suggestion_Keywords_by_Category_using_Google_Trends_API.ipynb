## Related Queries

It's a common thing that when a user searches for a topic, they would also search for something related. These are called related queries. Let us see what are the related queries for the topic "***Coronavirus***". Always remember when you want to change the topic name just run the following code again with the new name as the parameter.

```
import pandas as pd
from pytrends.request import TrendReq
```

```
pytrend = TrendReq(hl='en-US', tz=360)
```

```
# Create payload and capture API tokens. Only needed for interest_over_time(), interest_by_region() & related_queries()trendy_keywords = ‘Data Science’
kw_list = [''] #let it blank for keyword
#kw = trendy_keywords
pytrend.build_payload(kw_list, cat=184, #Business & Industrial: 12;Law & Government: 19; Jobs & Education: 958; Movies: 34; Autos & Vehicles: 47 ; Vehicle Brands: 815
                           timeframe='2022-07-01 2022-08-08', geo='MY', gprop='') #timeframe='now 7-d' (too short API cannot find); timeframe by Specific dates, '2016-12-14 2017-01-25'
```

Category listing ==> https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories

Google-Trends API ==> https://github.com/GeneralMills/pytrends#trending-searches


```
# Related Queries, returns a dictionary of dataframes
related_queries_dict = pytrend.related_queries()
#print(related_queries_dict)
rqueries = pd.DataFrame(related_queries_dict)
top = rqueries['']['top']
rising = rqueries['']['rising']

top.style.set_table_attributes("style='display:inline'").set_caption('Top Trending Keywords Table')
```

```
rising.style.set_table_attributes("style='display:inline'").set_caption('Rising Keywords Table')
```

```
# Related Topics, returns a dictionary of dataframes
related_topic = pytrend.related_topics()
#related_topic.values()
```

# Related Topics

```
topT = related_topic['']['top']
topT.style.set_table_attributes("style='display:inline'").set_caption('Top Related Topics')
```

```
risingT = related_topic['']['rising']
risingT.style.set_table_attributes("style='display:inline'").set_caption('Rising Related Topics')
```

## Keyword Suggestion
Google Trends can give you a list of keyword suggestions related to your primary keyword. In the example below, you will send a request to find suggestions for a keyword called “Business Intelligence.”

The suggestions method from pytrends will fetch keyword suggestions from Google Trends and return them in a data frame format.

```
keywords = pytrend.suggestions(keyword='honda hrv')
df = pd.DataFrame(keywords)
print(df)
```
