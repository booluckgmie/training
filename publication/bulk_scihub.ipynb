## Extract both in writing

```python
import re
import pandas as pd

def extract_dois(text):
    doi_pattern = re.compile(r'https://doi\.org/[\w\.\-\/]+')
    dois = doi_pattern.findall(text)
    return pd.DataFrame({'doi': dois})

def extract_semantic_links(text):
    ss_pattern = re.compile(r'https://semanticscholar\.org/[\w\.\-\/]+')
    ss_links = ss_pattern.findall(text)
    return pd.DataFrame({'ss_link': ss_links})

# text = '''Revised Reference with citation

# Introduction:
# Aging of the workforce is a global phenomenon, and many countries are facing the challenge of an aging population. In many developed countries, the proportion of the elderly in the workforce is rapidly increasing. As a result, there is growing interest in understanding the factors that influence the employment experience of older workers. In this literature review, we examined studies that have investigated the relationship between demographic factors and equality in the elderly workforce.
 
# Demographic Factors and Equality in Elderly Workforces
# Age is a critical demographic factor that influences the employment experiences of older workers. Studies have found that older workers face age-based discrimination, which includes negative attitudes, stereotypes, and perceptions of diminished capacity. For example, Petery et al.(2020) https://doi.org/10.1111/IJSA.12309 found that negative age stereotypes were associated with worse job performance among older workers. Moreover, age discrimination may lead to unequal opportunities for training, promotion, and career development, ultimately affecting the employment outcomes of older workers.
# https://doi.org/10.1007/s00391-019-01506-6
# https://doi.org/10.4236/PSYCH.2015.62012
# '''

# Extract DOIs
dois_df = extract_dois(text)
# print("DOIs:")
dois_df
```

```python
# Extract Semantic Scholar links
ss_links_df = extract_semantic_links(text)
# print("\nSemantic Scholar Links:")
ss_links_df
```

```python
paper_links = dois_df['doi'].tolist()
paper_links
```

## Download DOIs

```python
import requests
from scidownl import scihub_download

# paper_links = [
#     "https://doi.org/10.1145/3375633",
#     "https://doi.org/10.1109/5.771073",
#     "https://doi.org/10.1109/MCG.2005.156"
# ]

paper_type = "doi"
proxies = {
    'http': 'socks5://127.0.0.1:7890'
}

session = requests.Session()
session.proxies = proxies

for i, paper_link in enumerate(paper_links):
    out = f"./dois/paper_{i+1}.pdf"
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = session.get(f"https://sci-hub.se/{paper_link}", headers=headers)
    paper_url = response.url
    try:
        scihub_download(paper_url, paper_type=paper_type, out=out)
    except Exception as e:
        print(f"Could not download paper at {paper_link}: {e}")
```
