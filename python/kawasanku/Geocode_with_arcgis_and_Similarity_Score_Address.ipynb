```python
import pandas as pd 
from fuzzywuzzy import fuzz 
from fuzzywuzzy import process
```

```python
def checker(wrong_options, correct_options):
    names_array = []
    ratio_array = []
    for wrong_option in wrong_options:
        if wrong_option in correct_options:
            names_array.append(wrong_option)
            ratio_array.append("100")
        else:
            x = process.extractOne(
                wrong_option, correct_options, scorer=fuzz.token_set_ratio
            )
            names_array.append(x[0])
            ratio_array.append(x[1])
    return names_array, ratio_array
```

```python
original = df['full_address'].fillna('######').tolist()
suggest =geocoded_gdf['address'].fillna('######').tolist()
```

```python
name_match,ratio_match=checker(original,suggest)

df1 = pd.DataFrame()
df1['old_names']=pd.Series(original)
df1['correct_names']=pd.Series(name_match)
df1['correct_ratio']=pd.Series(ratio_match)
df1.to_excel('matched_names.xlsx')
```

```python
df1
```

```python
df1['QRatio'] = df1.apply(lambda row: fuzz.QRatio(row['old_names'], row['correct_names']), axis=1)
```

```python
df1[df1['QRatio']==81]
```
