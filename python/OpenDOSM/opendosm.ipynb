## lookup item

```python
# If not already installed, do: pip install pandas fastparquet

import pandas as pd

URL_DATA = 'https://storage.googleapis.com/dosm-public-pricecatcher/lookup_item.parquet'

lookup = pd.read_parquet(URL_DATA)
if 'date' in df.columns: df['date'] = pd.to_datetime(df['date'])

lookup.head()
```

## mapping df item

```python
df2 = df.merge(lookup, on='item_code', how='outer') 
# df2 = df2.dropna()
df2.shape
```

```python
df2.head()
```

```python
# df2 = df2.dropna()
df2 = df2.dropna(subset=['price'], axis=0)

df2.shape
```

```python
df2
```

```python
# Create a pivot table with date as the index, category as the columns and varience of price as the values
pivot_table = df2.pivot_table(values='price', index='date', columns='item_category', aggfunc='mean').reset_index() #var
```

```python
pivot_table
```

```python
pivot_table.columns
```

```python
df3 = pd.melt(pivot_table, id_vars=['date'], value_vars=['AYAM', 'DAGING', 'BUAH-BUAHAN', 'BAHAN LAUT', 'IKAN DARAT',
       'SAYUR-SAYURAN', 'KELAPA', 'TELUR', 'BAWANG', 'HASIL LAUT KERING',
       'KACANG', 'IKAN DALAM TIN', 'KICAP DAN SOS', 'TERSEDIA MINUM',
       'MINYAK DAN LEMAK', 'REMPAH RATUS (TIDAK BERBUNGKUS)',
       'REMPAH RATUS (BERBUNGKUS)', 'TEPUNG', 'UBI KENTANG',
       'MEE/KUETIAU', 'PENJAGAAN RUMAH', 'GULA', 'SANTAN (KOTAK)',
       'ESEN DAN RAGI', 'CILI KERING', 'TAUHU DAN TEMPE',
       'BAHAN-BAHAN MINUMAN', 'SAPUAN (SPREADS)', 'ROTI',
       'PENJAGAAN DIRI', 'SUSU BAYI', 'KRIMER DAN SUSU TEPUNG', 'BERAS',
       'MAKANAN BAYI', 'BIHUN', 'MENTEGA', 'LAMPIN PAKAI BUANG',
       'MI SEGERA'])
```

```python
df3 = df3.dropna()
df3
```
