```python
file_path = '2022-PENERBITAN BARU_1.sav'

# Read the SPSS data file
df, meta = pyreadstat.read_sav(file_path)

print(meta.column_names_to_labels)
```

```python
# Pretty print variable value labels
print("Variable Value Labels:")
for variable_name, value_labels in meta.variable_value_labels.items():
    print(f"{variable_name}:")
    for value, label in value_labels.items():
        print(f"  {value}: {label}")
```
