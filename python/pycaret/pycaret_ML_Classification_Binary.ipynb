```python
import pandas as pd
```

```python
data = pd.read_excel("/content/Customer-Data.xlsx")
```

```python
data
```

```python
from pycaret.classification import *
s = setup(data,target="Status", session_id=20)
```

```python
best =compare_models()
```

```python
evaluate_model(best)
```

```python
print(best)
```

```python
predict_model(best, data)
```

```python
training = predict_model(best, data)
```

```python
from pycaret.utils import check_metric
check_metric(training['Status'], training['Label'], metric = 'Accuracy')
```
