## Identify potential stopwords

```python
import pandas as pd
from collections import Counter

df = dictML

text = ' '.join(df['DESCRIPTION'].str.upper())
words = text.split()

## set of stopwords
excluded_stopwords = []

## count potential stopwords
counts = Counter(words)  

n = 3
top_3 = counts.most_common(n)

for word, count in top_3:
    if word in excluded_stopwords:
        top_3.remove((word, count))

#all data
df_all = pd.DataFrame.from_dict(counts, orient='index').rename(columns={0: 'count'})
df_all = df_all.reset_index()
df_all.sort_values('count', ascending=False, inplace=True)
df_all = df_all.rename(columns={'index': 'word'})

# potential stopwords top3
df_top = pd.DataFrame(top_3, columns=['word', 'count'])
df_top.sort_values('count', ascending=False, inplace=True)

print(df_all,'\n\n')
print(df_top,'\n\n')
```

## clean keywords column from any character, symbol

```python
df = dictML

# set of stopwords 
# excluded_stopwords = ['dan','di','sdn','bhd','&','dalam','atau','enterprise','lain','sdn.','bhd.','semua']
# excluded_stopwords = ['DAN','LAIN','ATAU','UNTUK','YANG','T.T.T.L.','DI','T.T.T.L','DARIPADA','DALAM','BAGI','DENGAN','MELALUI','SEMULA','BERKAITAN']
excluded_stopwords = ['DAN','LAIN','ATAU','UNTUK','YANG','T.T.T.L.','DI','T.T.T.L','DARIPADA','DALAM','BAGI','DENGAN','MELALUI','SEMULA','BERKAITAN',
                      'OF','AND','OTHER','N.E.C','OR','FOR','IN','THE','ON','BY','A','WITH','TO','RELATED','AS']
# excluded_stopwords = []

# loop through each row  
for index, row in df.iterrows():

    # split the text into words
    words = row['DESCRIPTION'].split() 

    # make words lowercase
    words = [word.upper() for word in words]

    # remove stopwords  
    words = [word for word in words if word not in excluded_stopwords]

    # convert "/" to " "
    words = [word.replace("/", " ") for word in words]

    # remove symbol or character
    words = [word for word in words if word.isalnum()]

    # update the row using .at[]   
    df.at[index, 'DESCRIPTION'] = ' '.join(words)

df['DESCRIPTION']
```

## SVM

```python
data = dictM2

# Import libraries
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

# Divide data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    [x[0] for x in data], [x[1] for x in data], test_size=0.2
)

# Create pipeline
pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SVC())  # Replace MultinomialNB with SVC
])

# Fit pipeline to training data
pipeline.fit(X_train, y_train)

# Predict test set labels
y_pred = pipeline.predict(X_test)

# Evaluate performance
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
