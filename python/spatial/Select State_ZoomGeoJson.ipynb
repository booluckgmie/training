#READ GEOJSON

```python
from pandas_geojson import read_geojson, filter_geojson
filter = ['P.005 Jerlun']
select = filter_geojson(geo_json=geo_json, filter_list=filter, property_key='parlimen')
```

```python
#print(select)
```

```python
import folium as fm
from folium import plugins

map = fm.Map(location = [5.016806622688496, 107.9456777002687], tiles='cartodbpositron', zoom_start = 6)

# Adds tool measure to the top right
#from folium.plugins import MeasureControl
#map.add_child(MeasureControl())

#map geojson
fm.GeoJson(select, name='DUN boundary',tooltip=fm.GeoJsonTooltip(fields=['dun', 'parlimen'])).add_to(map)
map.fit_bounds(map.get_bounds(), padding=(30, 30))
pn.panel(map, height=400)
```
