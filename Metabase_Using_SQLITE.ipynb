## Convert CSV to SQLite File

```python
import csv_to_sqlite 

# all the usual options are supported
options = csv_to_sqlite.CsvOptions(typing_style="full", encoding="windows-1250") 
input_files = ["/content/sample_data/california_housing_test.csv"] # pass in a list of CSV files
csv_to_sqlite.write_csv(input_files, "housing.sqlite", options)
```

## Read SQLite File

```python
import pandas as pd
import sqlite3
from google.colab import data_table
data_table.enable_dataframe_formatter()

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("housing.sqlite")
cursor = con.cursor()

#fetch table in this database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
#read data in table 
df = pd.read_sql_query("SELECT * from california_housing_test", con)

# Verify that result of SQL query is stored in the dataframe
df

#con.close()
```
